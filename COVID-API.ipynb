```python
import requests
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
response = requests.get('https://api.covid19api.com/summary')
print(response)
```

```python
summary_string = response.content
summary = json.loads(summary_string)
```

```python
summary
```

```python
countries = summary["Countries"]
countries
```

```python
countries_json = json.dumps(countries)
df = pd.read_json(countries_json)
print(df)
df
```

```python
df_country = df.set_index("Country")
```

```python
df_country
```

```python
state = "Israel"
df_country.loc[state,:]
```

```python
response = requests.get("https://api.covid19api.com/total/country/Israel")
print(response)
```

```python
contents = json.loads(response.content)
```

```python
contents
```

```python
df_israel = pd.read_json(response.content)
```

```python
df_israel
shape = np.shape(df_israel)
shape[0]
```

```python
conf = df_israel.loc[shape[0]-70:shape[0],"Confirmed"]
conf
```

```python
plt.plot(conf)
plt.title("Confirmed - daily")
plt.xlabel("Days from outbreak [day]")
plt.ylabel("Confirmed sick [person]")
```

```python
df_israel
```

```python
recovered = df_israel.loc[shape[0]-70:shape[0]:,"Recovered"]
```

```python
recovered
```

```python
active = conf - recovered
```

```python
active
```

```python
plt.plot(active)
plt.title("Active - Daily")
plt.xlabel("Days from outbreak [day]")
plt.ylabel("Active sick [person]")
```

```python
active_diff = active.diff()
active_diff[active_diff < 0]
plt.plot(active_diff)
```
